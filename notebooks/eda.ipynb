```python
import matplotlib.pyplot as plt
import seaborn as sns #You should also import seaborn
import pandas as pd #Import pandas as well
```

```python
df = pd.read_csv("../src/Data/MachineLearningRating_v3.csv" ,delimiter='|')
df.head()
```

```python
df.describe()
```

```python
# Check data types of all columns
print("Data Types:\n", df.dtypes)

# Convert any necessary columns (e.g., `TransactionMonth` to datetime)
df['TransactionMonth'] = pd.to_datetime(df['TransactionMonth'])
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print("Missing Values:\n", missing_values)
```

```python
# Histograms for numerical columns
numerical_cols = ['TotalPremium', 'TotalClaims', 'SumInsured']

df[numerical_cols].hist(bins=30, figsize=(12, 8))
plt.suptitle('Distribution of Numerical Columns')
plt.show()
```

```python
# Bar charts for categorical columns
categorical_cols = ['Gender', 'Province', 'CoverType']

for col in categorical_cols:
    plt.figure(figsize=(8, 5))
    sns.countplot(data=df, x=col)
    plt.title(f'{col} Distribution')
    plt.show()
```

```python
# Scatter plot for TotalPremium vs TotalClaims
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='TotalPremium', y='TotalClaims')
plt.title('TotalPremium vs TotalClaims')
plt.show()
```

```python
# Correlation matrix heatmap
corr_matrix = df[numerical_cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Box plots to detect outliers in numerical columns
for col in numerical_cols:
    plt.figure(figsize=(8, 5))
    sns.boxplot(data=df, x=col)
    plt.title(f'{col} Boxplot')
    plt.show()
```
